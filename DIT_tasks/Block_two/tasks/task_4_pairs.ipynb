```python
#Исходные данные
items = [
    (1, 'Apple', 'Fruit'),
    (2, 'Banana', 'Fruit'),
    (3, 'Apple', 'Fruit'),
    (4, 'Orange', 'Fruit'),
    (5, 'Banana', 'Fruit'),
    (6, 'Carrot', 'Vegetable'),
    (7, 'Carrot', 'Vegetable'),
    (8, 'Apple', 'Vegetable'),
    (9, 'Laptop', 'Electronics'),
    (10, 'Phone', 'Electronics'),
    (11, 'Laptop', 'Electronics'),
    (12, 'Laptop', 'Electronics'),
]

#Создаем словарь для группировки
groups = {}

#Проходим по всем записям
for item in items:
    id = item[0]
    name = item[1]
    category = item[2]

    key = (name, category)

    if key not in groups:
        groups[key] = []

    groups[key].append(id)


#Создаем список для результата
result = []

for key, ids in groups.items():
    name = key[0]
    category = key[1]

    if len(ids) >= 2:
        for i in range(len(ids)):
            for j in range(len(ids)):
                if i == j:
                    continue

                id1 = ids[i]
                id2 = ids[j]

                if id1 < id2:
                    pair = (id1, id2, category)
                    result.append(pair)

print("id1 | id2 | category")

result.sort()

for pair in result:
    id1 = pair[0]
    id2 = pair[1]
    category = pair[2]
    print(f"{id1:3} | {id2:3} | {category}")

print()
print(f"Всего найдено пар: {len(result)}")
```
