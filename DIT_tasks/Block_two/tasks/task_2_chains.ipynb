```python
#Итеративный подход (без рекурсии)
def solve_iterative():
    # Исходные данные
    users = [(1, 2), (2, 3), (3, 5), (4, 4), (7, 8), (8, 7)]

    #Создаём словарь связей: id -> new_id
    next_id = {}
    for id, new_id in users:
        next_id[id] = new_id

    #Собираем все уникальные ID
    all_ids = []
    for id, new_id in users:
        if id not in all_ids:
            all_ids.append(id)
        if new_id not in all_ids:
            all_ids.append(new_id)

    #Для каждого ID находим минимальный в цепочке
    result = {}
    for start_id in all_ids:
        current_id = start_id
        chain = [current_id]

        #Идём по цепочке до конца
        while current_id in next_id:
            next_val = next_id[current_id]
            if next_val in chain:
                break
            chain.append(next_val)
            current_id = next_val

        #Находим минимальный ID в цепочке
        min_id = min(chain)
        result[start_id] = min_id

    return result


#Рекурсивная функция для поиска минимального ID
def find_min_recursive(current_id, next_id, visited):
    visited.append(current_id)

    #Если есть следующий ID и мы его ещё не посещали
    if current_id in next_id:
        next_val = next_id[current_id]
        if next_val not in visited:
            return find_min_recursive(next_val, next_id, visited)

    #Возвращаем минимальный из всех посещённых
    return min(visited)


#Рекурсивный подход
def solve_recursive():
    users = [(1, 2), (2, 3), (3, 5), (4, 4), (7, 8), (8, 7)]

    #Создаём словарь связей
    next_id = {}
    for id, new_id in users:
        next_id[id] = new_id

    #Собираем все уникальные ID
    all_ids = []
    for id, new_id in users:
        if id not in all_ids:
            all_ids.append(id)
        if new_id not in all_ids:
            all_ids.append(new_id)

    #Для каждого ID находим минимальный рекурсивно
    result = {}
    for id in all_ids:
        visited = []
        min_id = find_min_recursive(id, next_id, visited)
        result[id] = min_id

    return result

result_iter = solve_iterative()
result_rec = solve_recursive()

#Выводим результаты
print("Iterative approach:")
for id in sorted(result_iter.keys()):
    print(f"{id} -> {result_iter[id]}")

print("\nRecursive approach:")
for id in sorted(result_rec.keys()):
    print(f"{id} -> {result_rec[id]}")

#Группируем по конечным ID
chains = {}
for id in sorted(result_iter.keys()):
    final_id = result_iter[id]
    if final_id not in chains:
        chains[final_id] = []
    chains[final_id].append(id)

print("\nGroups:")
for final_id in sorted(chains.keys()):
    print(f"{chains[final_id]} -> {final_id}")
```
